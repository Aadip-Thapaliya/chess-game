```python
def readable_board(func):
    def wrapper():
        x=func()
        for i in func:
            print(i)
        return x
    return wrapper

@readable_board
def board_initialize():
        chess_board=[["wc","wknight","wb","wq","wking","wb","wknight","wc"],
                ["wp"]*8,
                [None]*8,
                [None]*8,
                [None]*8,
                [None]*8,
                ["bp"]*8,
                ["bc","bknight","bb","bq","bking","bb","bknight","bc"]]
        return chess_board
board_initialize()
```
